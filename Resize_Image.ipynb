```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/687px-Mona_Lisa,_by_Leonardo_da_Vinci,_from_C2RMF_retouched.jpg')
cv2_imshow(image)
# percent by which the image is resized
scale_percent = 50
# calculate 50% of original dimension
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)

# dsize
dsize = (width,height)

# resize image
half = cv2.resize(image, dsize)
cv2_imshow(half)
```
